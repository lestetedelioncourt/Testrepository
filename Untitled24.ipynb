```python
def get_XT_using_ddim(Xt, diffusion_model, device=device):
  time_steps = torch.flip(diffusion_model.scheduler.timesteps, dims=(0,))
  t = time_steps[:-1]
  t_next = time_steps[1:]

  for n, (i, j) in enumerate(zip(t.tolist(), t_next.tolist())):
    model_input = diffusion_model.scheduler.scale_model_input(Xt, i).to(device)

    with torch.no_grad():
      pred_epsilon_t = diffusion_model.unet(model_input, i)["sample"]

    alpha_bar_t_plus_delta_t = diffusion_model.scheduler.alphas_cumprod[j]
    alpha_bar_t = diffusion_model.scheduler.alphas_cumprod[i]

    pred_X0 = (Xt - pred_epsilon_t * (1 - alpha_bar_t).sqrt()) / alpha_bar_t.sqrt()
    #pred_X0 = pred_X0.clamp(-1, 1)

    X_t_plus_delta_t = alpha_bar_t_plus_delta_t.sqrt() * pred_X0 + (1 - alpha_bar_t_plus_delta_t).sqrt() * pred_epsilon_t

    Xt = X_t_plus_delta_t

  return Xt
```

```python
h_content = None
h_style = None
gate = None
def get_h_content(module, input, output):
  global h_content
  global gate
  if gate == "content":
    h_content = output
  h_content_hook.remove()

def get_h_style(module, input, output):
  global h_style
  global gate

  if gate == "style":
    h_style = output
  h_style_hook.remove()

def combine_style_and_content(module, input, output):
  global gamma
  global h_style
  global h_content
  global gate
  if gate == "combine":
    h_hat_hook.remove()
    return (1-gamma) * h_style + gamma * h_content
```

```python
timesteps = image_pipe.scheduler.timesteps
X_t_style = Xt_style
X_t_content = XT_content

X_t_new = None
gamma = 0.7 # best value is usually 0.6, but when working with out of domain images should experiment with value
edit_steps = 400
for i, t in enumerate(image_pipe.scheduler.timesteps):
  alpha_bar_t = image_pipe.scheduler.alphas_cumprod[t]
  if t != 0:
    alpha_bar_t_minus_delta_t = image_pipe.scheduler.alphas_cumprod[image_pipe.scheduler.timesteps[i + 1]]
  else:
    alpha_bar_t_minus_delta_t = torch.tensor([1]).to(device)

  if t > edit_steps:
    # get h content
    gate = "content"
    model_input = image_pipe.scheduler.scale_model_input(X_t_content, t)
    h_content_hook = image_pipe.unet.down_blocks[5].resnets[1].conv2.register_forward_hook(get_h_content)
    # final downblock, resnet block index 1, convolution layer index 2, will use to get content feature maps from unet
    # pytorch will call get_h_content automatically at model bottleneck (where we have specified)

    with torch.no_grad():
      pred_epsilon = image_pipe.unet(model_input, t)["sample"]

    predicted_X0 = (X_t_content - (1 - alpha_bar_t).sqrt() * pred_epsilon) / alpha_bar_t.sqrt()
    # predicted_X0 = predicted_X0.clamp(-image_pipe.scheduler.config.clip_sample_range, image_pipe.scheduler.config.clip_sample_range)
    # should not restrict predicted_X0 when injecting content to get the best predictions

    X_t_minus_delta_t = alpha_bar_t_minus_delta_t.sqrt() * predicted_X0 + (1 - alpha_bar_t_minus_delta_t) * pred_epsilon
    X_t_content = X_t_minus_delta_t

    # get h style
    gate = "style"

    h_style_hook = image_pipe.unet.down_blocks[5].resnets[1].conv2.register_forward_hook(get_h_style)
    # will take style layer from the same layer as h_content tensor

    if i == 0:
      # if at first timestep (timestep T)
      model_input = image_pipe.scheduler.scale_model_input(Xt_style, t)
      with torch.no_grad():
        image_pipe.unet(model_input, t)["sample"]
    else:
      model_input = image_pipe.scheduler.scale_model_input(X_t_new, t)
      with torch.no_grad():
        image_pipe.unet(model_input, t)["sample"]

    # combine style and content
    gate = None
    with torch.no_grad():
      if i == 0:
        model_input = image_pipe.scheduler.scale_model_input(Xt_style, t)
      else:
        model_input = image_pipe.scheduler.scale_model_input(X_t_new, t)
      pred_epsilon_given_h_t = image_pipe.unet(model_input, t)["sample"]

    gate = "combine"
    h_hat_hook = image_pipe.unet.down_blocks[5].resnets[1].conv2.register_forward_hook(combine_style_and_content)
    if i == 0:
      with torch.no_grad():
        model_input = image_pipe.scheduler.scale_model_input(Xt_style, t)
        pred_epsilon_given_h_hat = image_pipe.unet(model_input, t)["sample"]

      predicted_X0 = (Xt_style - (1 - alpha_bar_t).sqrt() * pred_epsilon_given_h_hat) / alpha_bar_t.sqrt()
      # predicted_X0 = predicted_X0.clamp(-image_pipe.scheduler.config.clip_sample_range, image_pipe.scheduler.config.clip_sample_range)

      X_t_minus_delta_t = alpha_bar_t_minus_delta_t.sqrt() * predicted_X0 + (1 - alpha_bar_t_minus_delta_t).sqrt() * pred_epsilon_given_h_t
      X_t_new = X_t_minus_delta_t
    else:
      with torch.no_grad():
        model_input = image_pipe.scheduler.scale_model_input(X_t_new, t)
        pred_epsilon_given_h_hat = image_pipe.unet(model_input, t)["sample"]
      predicted_X0 = (X_t_new - (1 - alpha_bar_t).sqrt() * pred_epsilon_given_h_hat) / alpha_bar_t.sqrt()
      # predicted_X0 = predicted_X0.clamp(-image_pipe.scheduler.config.clip_sample_range, image_pipe.scheduler.config.clip_sample_range)

      X_t_minus_delta_t = alpha_bar_t_minus_delta_t.sqrt() * predicted_X0 + (1 - alpha_bar_t_minus_delta_t).sqrt() * pred_epsilon_given_h_t
      X_t_new = X_t_minus_delta_t
  else:
    with torch.no_grad():
      model_input = image_pipe.scheduler.scale_model_input(X_t_new, t)
      pred_epsilon = image_pipe.unet(model_input, t)["sample"]
    predicted_X0 = (X_t_new - (1 - alpha_bar_t).sqrt() * pred_epsilon) / alpha_bar_t.sqrt()
    predicted_X0 = predicted_X0.clamp(-image_pipe.scheduler.config.clip_sample_range, image_pipe.scheduler.config.clip_sample_range)
    # clamp used here becaause not doing content injection

    X_t_minus_delta_t = alpha_bar_t_minus_delta_t.sqrt() * predicted_X0 + (1 - alpha_bar_t_minus_delta_t).sqrt() * pred_epsilon
    X_t_new = X_t_minus_delta_t

```
