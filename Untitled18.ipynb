```python
import numpy as np
import torch
import torch.nn.functional as F
from matplotlib import pyplot as plt
from PIL import Image
from torchvision import datasets
```

```python
!pip install datasets
from datasets import load_dataset
dataset = load_dataset("huggan/smithsonian_butterflies_subset", split="train")
```

```python
from torchvision import transforms
```

```python
image_size = 128

preprocess = transforms.Compose(
    [
        transforms.Resize((image_size, image_size)),
        transforms.RandomHorizontalFlip(),
        #transforms.RandomRotation(degrees=40),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5]), # normalizes to range between -1 and 1
    ]
)
```

```python
def transform(examples):
  images = [preprocess(image.convert("RGB")) for image in examples["image"]] # re-orders BGR to RGB, applys transformation
  return {"images": images}
```

```python
dataset.set_transform(transform)
```

```python
batch_size = 16

train_dataloader = torch.utils.data.DataLoader(
    dataset, batch_size=batch_size, shuffle=True
)
```

```python
from diffusers import UNet2DModel
device = "cuda" if torch.cuda.is_available() else "cpu"
# Create a model
model = UNet2DModel(
    sample_size = image_size, # The target image resolution
    in_channels=3, # the number of input channels, 3 for RGB images
    out_channels=3, # the number of output channels
    layers_per_block=2, # how many ResNet layers to use per UNet block
    block_out_channels=(128, 128, 256, 256, 512, 512), # the number of output channels for each UNet block
    down_block_types=(
        "DownBlock2D", # a regular ResNet downsampling block # out_channels = 128
        "DownBlock2D", # out_channels = 128
        "DownBlock2D", # out_channels = 256
        "DownBlock2D", # out_channels = 256
        "AttnDownBlock2D", # a ResNet downsampling block with spatial self-attention # out_channels = 512
        "DownBlock2D", # out_channels = 512
    ),
    up_block_types=(
        "UpBlock2D", # a regular ResNet upsampling block # out_channels = 512
        "AttnUpBlock2D", # a ResNet upsampling block with spatial self-attention # out_channels = 512
        "UpBlock2D", # out_channels = 256
        "UpBlock2D", # out_channels = 256
        "UpBlock2D", # out_channels = 128
        "UpBlock2D" # out_channels = 128
    ),
)
model.to(device)
```

```python
# Set the noise scheduler
from diffusers import DDPMScheduler

# Will use to add noise to the image
noise_scheduler = DDPMScheduler(
    num_train_timesteps=1000, beta_schedule="squaredcos_cap_v2"
)
```

```python

# Training loop
optimizer = torch.optim.AdamW(model.parameters(), lr=4e-4)

losses = []

for epoch in range(100):
  for step, batch in enumerate(train_dataloader):
    print(step)
    clean_images = batch["images"].to(device)
    noise = torch.randn(clean_images.shape).to(clean_images.device)
    bs = clean_images.shape[0] # batch size

    # create timesteps for each image in batch
    timesteps = torch.randint(
        0, noise_scheduler.num_train_timesteps, (bs,), device=clean_images.device
    ).long()

    noisy_images = noise_scheduler.add_noise(clean_images, noise, timesteps)

    noise_pred = model(noisy_images, timesteps, return_dict=False)[0]

    loss = F.mse_loss(noise_pred, noise)
    loss.backward()
    losses.append(loss.item())

    optimizer.step()
    optimizer.zero_grad()

  if (epoch + 1) % 5 == 0:
    loss_last_epoch = sum(losses[-len(train_dataloader) :]) / len(train_dataloader)
    print(f"Epoch:{epoch+1}, loss: {loss_last_epoch}")
```
