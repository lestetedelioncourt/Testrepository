```python
def get_slice_bboxes(
    image_height: int,
    image_width: int,
    slice_height: Optional[int] = None,
    slice_width: Optional[int] = None,
    auto_slice_resolution: bool = True,
    overlap_height_ratio: float = 0.0,
    overlap_width_ratio: float = 0.0,
) -> List[List[int]]:

  slice_bboxes = []
  y_max = y_min = 0

  if slice_height and slice_width:
    y_overlap = int(overlap_height_ratio * slice_height)
    x_overlap = int(overlap_width_ratio * slice_width)
  else:
    raise ValueError("Compute type is not auto and slice width and height are not provided.")

  while y_max < image_height:
    x_min = x_max = 0
    y_max = y_min + slice_height
    while x_max < image_width:
      x_max = x_min + slice_width
      if y_max > image_height or x_max > image_width:
        xmax = min(image_width, x_max)
        ymax = min(image_height, y_max)
        xmin = max(0, xmax - slice_width)
        ymin = max(0, ymax - slice_height)
        slice_bboxes.append([xmin, ymin, xmax, ymax])
      else:
        slice_bboxes.append([x_min, y_min, x_max, y_max])
      x_min = x_max - x_overlap
    y_min = y_max - y_overlap
  return slice_bboxes
```

```python
class SlicedImage:
  def __init__(self, image, starting_pixel):
      self.image = image
      self.starting_pixel = starting_pixel
```

```python
class SliceImageResult:
  def __init__(self, original_image_size: List[int], image_dir: Optional[str] = None):
    self.original_image_height = original_image_size[0]
    self.original_image_width = original_image_size[1]
    self.image_dir = image_dir

    self._sliced_image_list: List[SlicedImage] = []

  def add_sliced_image(self, sliced_image: SlicedImage):
    if not isinstance(sliced_image, SlicedImage):
        raise TypeError("sliced_image must be a SlicedImage instance")

    self._sliced_image_list.append(sliced_image)

  @property
  def sliced_image_list(self):
    return self._sliced_image_list

  @property
  def starting_pixels(self) -> List[int]:
    starting_pixels = []
    for sliced_image in self._sliced_image_list:
      starting_pixels.append(sliced_image.starting_pixels)
    return starting_pixels

  @property
  def filenames(self) -> List[int]:
    filenames = []
    for sliced_image in self._sliced_image_list:
      filenames.append(sliced_image.coco_image.file_name)
    return filenames

  def __getitem__(self, i):
    def _prepare_ith_dict(i):
      return {
          "image": self.images[i],
          "starting_pixel": self.starting_pixels[i],
      }

    if isinstance(i, np.ndarray):
      i = i.tolist()

    if isinstance(i, int):
      return _prepare_ith_dict(i)
    elif isinstance(i, slice):
      start, stop, step = i.indices(len(self))
      return [_prepare_ith_dict(i) for i in range(start, stop, step)]
    elif isinstance(i, (tuple, list)):
      accessed_mapping = map(_prepare_ith_dict, i)
      return list(accessed_mapping)
    else:
      raise NotImplementedError(f"{type(i)}")

  def __len__(self):
    return len(self._sliced_image_list)
```

```python
def slice_image_func(
    image: Union[str, Image.Image],
    slice_height: Optional[int] = None,
    slice_width: Optional[int] = None,
    overlap_height_ratio: float = 0.0,
    overlap_width_ratio: float = 0.0,
    auto_slice_resolution: bool = True,
    min_area_ratio: float = 0.1,
    out_ext: Optional[str] = None,
    verbose: bool = False,
) -> SliceImageResult:

  image_pil = image

  image_width, image_height = image_pil.size
  if (image_width == 0 and image_height == 0):
      raise RuntimeError(f"invalid image size: {image_pil.size} for 'slice_image'.")
  slice_bboxes = get_slice_bboxes(
      image_height = image_height,
      image_width = image_width,
      auto_slice_resolution = auto_slice_resolution,
      slice_height = slice_height,
      slice_width = slice_width,
      overlap_height_ratio = overlap_height_ratio,
      overlap_width_ratio = overlap_width_ratio
  )

  n_ims = 0

  sliced_image_result = SliceImageResult(original_image_size=[image_height, image_width])

  image_pil_arr = np.asarray(image_pil)

  for slice_bbox in slice_bboxes:
    n_ims += 1

    tlx = slice_bbox[0] # top left
    tly = slice_bbox[1]
    brx = slice_bbox[2] # bottom right
    bry = slice_bbox[3]
    image_pil_slice = image_pil_arr[tly:bry, tlx:brx]

    slice_width = brx - tlx
    slice_height = bry - tly

    sliced_image = SlicedImage(
        image=image_pil_slice, starting_pixel=[slice_bbox[0], slice_bbox[1]]
    )
    sliced_image_result.add_sliced_image(sliced_image)

  return sliced_image_result
```

```python
def slice_image(
    image_path,
    slice_height,
    slice_width,
    overlap_height_ratio,
    overlap_width_ratio):
  image = Image.open(image_path)
  image_width, image_height = image.size
  slice_bboxes = get_slice_bboxes(
      image_height,
      image_width,
      slice_height,
      slice_width,
      False,
      overlap_height_ratio,
      overlap_width_ratio
  )

  slices = []
  for bbox in slice_bboxes:
    slice = image.crop(bbox)
    slices.append((slice, bbox))
  return slices
```

```python
def adjust_annotations(annotations, x_min, y_min, slice_width, slice_height, image_width, image_height):
    adjusted_annotations = []
    for ann in annotations:
        class_id, x_center, y_center, width, height = ann

        # Convert normalized coordinates to pixel values
        x_center_px = x_center * image_width
        y_center_px = y_center * image_height
        width_px = width * image_width
        height_px = height * image_height

        # Adjust to slice coordinates
        x_center_slice = x_center_px - x_min
        y_center_slice = y_center_px - y_min

        # Check if the bounding box intersects with the slice
        if (x_center_slice + width_px/2 > 0 and x_center_slice - width_px/2 < slice_width and
            y_center_slice + height_px/2 > 0 and y_center_slice - height_px/2 < slice_height):

            # Clip the bounding box to the slice boundaries
            x_min_slice = max(0, x_center_slice - width_px/2)
            y_min_slice = max(0, y_center_slice - height_px/2)
            x_max_slice = min(slice_width, x_center_slice + width_px/2)
            y_max_slice = min(slice_height, y_center_slice + height_px/2)

            # Calculate new center and dimensions
            new_width = x_max_slice - x_min_slice
            new_height = y_max_slice - y_min_slice
            new_x_center = (x_min_slice + x_max_slice) / 2
            new_y_center = (y_min_slice + y_max_slice) / 2

            # Normalize to slice dimensions
            new_x_center /= slice_width
            new_y_center /= slice_height
            new_width /= slice_width
            new_height /= slice_height

            adjusted_annotations.append([class_id, new_x_center, new_y_center, new_width, new_height])

    return adjusted_annotations
```

```python
def process_dataset(data_yaml, output_dir):
  with open(data_yaml, 'r') as file:
    config = yaml.safe_load(file)

  image_width  = 5472
  image_height = 3648
  slice_height, slice_width = config['tile_size']
  overlap_height, overlap_width = config['tile_overlap']
  overlap_height_ratio = overlap_height / slice_height
  overlap_width_ratio = overlap_width / slice_width
  max_tiles = config['max_tiles']

  for split in ['train', 'val']:
    images_dir = os.path.join(os.path.dirname(data_yaml), config[split])
    labels_dir = os.path.join(os.path.dirname(data_yaml), 'labels', split)
    output_images_dir = os.path.join(output_dir, 'images', split)
    output_labels_dir = os.path.join(output_dir, 'labels', split)

    os.makedirs(output_images_dir, exist_ok=True)
    os.makedirs(output_labels_dir, exist_ok=True)

    j=0

    for image_file in tqdm(os.listdir(images_dir), desc=f"Processing {split} set"):
      j+=1
      image_path = os.path.join(images_dir, image_file)
      label_path = os.path.join(labels_dir, os.path.splitext(image_file)[0] + '.txt')

      slices = slice_image(image_path,
                           slice_height,
                           slice_width,
                           overlap_height_ratio,
                           overlap_width_ratio)

      if os.path.exists(label_path):
        with open(label_path, 'r') as f:
          annotations = [list(map(float, line.strip().split())) for line in f]
      else:
        annotations = []

      for i, (slice_img, bbox) in enumerate(slices[:max_tiles]):
        x_min, y_min, x_max, y_max = bbox
        adjusted_anns = adjust_annotations(annotations,
                                           x_min,
                                           y_min,
                                           x_max - x_min,
                                           y_max - y_min,
                                           image_width,
                                           image_height)

        if adjusted_anns:
          #slice_name = f"{os.path.splitext(image_file[0])}_tile-{i}.png"
          slice_name = f"tile_{j}_{i}.png"
          slice_img.save(os.path.join(output_images_dir, slice_name))

          with open(os.path.join(output_labels_dir, f"tile_{j}_{i}.txt"), 'w') as f:
            for ann in adjusted_anns:
              f.write(' '.join(map(str, ann)) + '\n')
```

```python
import cv2
import numpy as np
from PIL import Image
from ultralytics import YOLO
import torch

def inference_with_slice_boxing(model_path, image_path, image_save_path, annotation_save_path,
                                slice_height=114, slice_width=114,
                                overlap_height_ratio=0.0, overlap_width_ratio=0.0,
                                conf_threshold=0.06, iou_threshold=0.45):
    # Load the model
    model = YOLO(model_path)

    # Load and prepare the image
    image = Image.open(image_path)
    image_width, image_height = image.size
    image_np = np.array(image)

    # Get slice bboxes
    slice_bboxes = get_slice_bboxes(
        image_height=image_height,
        image_width=image_width,
        slice_height=slice_height,
        slice_width=slice_width,
        auto_slice_resolution=False,
        overlap_height_ratio=overlap_height_ratio,
        overlap_width_ratio=overlap_width_ratio
    )

    # Initialize lists to hold detections
    all_boxes = []
    all_scores = []
    all_classes = []

    # Perform inference on each slice
    for bbox in slice_bboxes:
        x_min, y_min, x_max, y_max = bbox
        slice_img = image_np[y_min:y_max, x_min:x_max]

        # Perform inference on the slice
        results = model(slice_img, conf=conf_threshold, iou=iou_threshold)

        # Adjust bounding boxes to full image coordinates
        for r in results:
            boxes = r.boxes.xyxy  # Absolute coordinates in the slice
            confs = r.boxes.conf
            classes = r.boxes.cls  # If classes are needed

            for box, conf, cls in zip(boxes, confs, classes):
                x1, y1, x2, y2 = box.cpu().numpy()
                conf = conf.cpu().item()
                cls = int(cls.cpu().item())

                # Adjust coordinates to full image
                x1_abs = x1 + x_min
                y1_abs = y1 + y_min
                x2_abs = x2 + x_min
                y2_abs = y2 + y_min

                # Ensure the coordinates are within image boundaries
                x1_abs = max(0, min(x1_abs, image_width - 1))
                y1_abs = max(0, min(y1_abs, image_height - 1))
                x2_abs = max(0, min(x2_abs, image_width - 1))
                y2_abs = max(0, min(y2_abs, image_height - 1))

                # Append to list
                all_boxes.append([x1_abs, y1_abs, x2_abs, y2_abs])
                all_scores.append(conf)
                all_classes.append(cls)

    # Convert lists to tensors
    if len(all_boxes) > 0:
        all_boxes_tensor = torch.tensor(all_boxes, dtype=torch.float32)
        all_scores_tensor = torch.tensor(all_scores, dtype=torch.float32)
        all_classes_tensor = torch.tensor(all_classes, dtype=int)

        # Apply NMS
        indices = nms(all_boxes_tensor, all_scores_tensor, iou_threshold)

        # Get final detections
        final_boxes = all_boxes_tensor[indices].numpy()
        final_scores = all_scores_tensor[indices].numpy()
        final_classes = all_classes_tensor[indices].numpy()
    else:
        final_boxes = np.array([])
        final_scores = np.array([])
        final_classes = np.array([])

    # Annotate the image
    image_with_boxes = image_np.copy()
    if len(final_boxes) > 0:
        for box in final_boxes:
            x1, y1, x2, y2 = box
            x1 = int(x1)
            y1 = int(y1)
            x2 = int(x2)
            y2 = int(y2)
            cv2.rectangle(image_with_boxes, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Save annotated image
    cv2.imwrite(image_save_path, cv2.cvtColor(image_with_boxes, cv2.COLOR_RGB2BGR))

    # Save annotations in YOLO format
    with open(annotation_save_path, 'w') as f:
        for box, cls in zip(final_boxes, final_classes):
            x1, y1, x2, y2 = box
            x_center = ((x1 + x2) / 2) / image_width
            y_center = ((y1 + y2) / 2) / image_height
            width = (x2 - x1) / image_width
            height = (y2 - y1) / image_height
            line = f"{cls} {x_center} {y_center} {width} {height}\n"
            f.write(line)

    return final_boxes, image_with_boxes, final_scores
```

```python
import cv2
import numpy as np
from PIL import Image
from ultralytics import YOLO
import torch
from torchvision.ops import nms
import itertools
from concurrent.futures import ThreadPoolExecutor

def inference_with_slice_boxing3(model_path, image_path, image_save_path, annotation_save_path,
                                slice_heights, slice_widths,
                                overlap_height_ratio=0.1, overlap_width_ratio=0.1,
                                conf_threshold=0.15, iou_threshold=0.45, max_workers=4):
    # Load the model
    model = YOLO(model_path)

    # Load and prepare the image
    image = Image.open(image_path)
    image_width, image_height = image.size
    image_np = np.array(image)

    # Initialize list to hold detections
    all_detections = []

    # Function to process a single slice
    def process_slice(args):
        slice_img, x_min, y_min, slice_width, slice_height = args
        results = model(slice_img, conf=conf_threshold, iou=iou_threshold)
        detections = []
        for r in results:
            boxes = r.boxes.xyxy
            confs = r.boxes.conf
            classes = r.boxes.cls
            if len(boxes) == 0:
                continuel
            boxes = boxes.cpu().numpy()
            confs = confs.cpu().numpy()
            classes = classes.cpu().numpy()

            # Adjust coordinates
            x1_abs = (boxes[:, 0] + x_min) / image_width
            y1_abs = (boxes[:, 1] + y_min) / image_height
            x2_abs = (boxes[:, 2] + x_min) / image_width
            y2_abs = (boxes[:, 3] + y_min) / image_height

            x_centers = (x1_abs + x2_abs) / 2
            y_centers = (y1_abs + y2_abs) / 2
            widths = x2_abs - x1_abs
            heights = y2_abs - y1_abs

            # Stack detections
            detections_slice = np.stack((classes, x_centers, y_centers, widths, heights, confs), axis=1)
            detections.append(detections_slice)
        if detections:
            return np.concatenate(detections, axis=0)
        else:
            return np.array([])

    # Iterate over all combinations of slice heights and widths
    for slice_height, slice_width in itertools.product(slice_heights, slice_widths):
        # Get slice bboxes
        slice_bboxes = get_slice_bboxes(
            image_height=image_height,
            image_width=image_width,
            slice_height=slice_height,
            slice_width=slice_width,
            auto_slice_resolution=False,
            overlap_height_ratio=overlap_height_ratio,
            overlap_width_ratio=overlap_width_ratio
        )

        # Prepare arguments for parallel processing
        args_list = []
        for bbox in slice_bboxes:
            x_min, y_min, x_max, y_max = bbox
            slice_img = image_np[y_min:y_max, x_min:x_max]
            args_list.append((slice_img, x_min, y_min, slice_width, slice_height))

        # Process slices in parallel
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            results_list = list(executor.map(process_slice, args_list))

        # Collect detections
        for detections in results_list:
            if detections.size > 0:
                all_detections.append(detections)

    if all_detections:
        # Concatenate all detections
        all_detections = np.concatenate(all_detections, axis=0)
        # Extract boxes and scores for NMS
        boxes = all_detections[:, 1:5]
        scores = all_detections[:, 5]
        classes = all_detections[:, 0].astype(int)

        # Convert boxes to absolute coordinates for NMS
        boxes_abs = np.zeros_like(boxes)
        boxes_abs[:, 0] = (boxes[:, 0] - boxes[:, 2] / 2) * image_width  # x1
        boxes_abs[:, 1] = (boxes[:, 1] - boxes[:, 3] / 2) * image_height  # y1
        boxes_abs[:, 2] = (boxes[:, 0] + boxes[:, 2] / 2) * image_width  # x2
        boxes_abs[:, 3] = (boxes[:, 1] + boxes[:, 3] / 2) * image_height  # y2

        # Convert to tensors
        boxes_tensor = torch.tensor(boxes_abs)
        scores_tensor = torch.tensor(scores)
        classes_tensor = torch.tensor(classes)

        # Perform NMS
        indices = nms(boxes_tensor, scores_tensor, iou_threshold)

        # Get final detections
        final_boxes = boxes[indices]
        final_scores = scores[indices]
        final_classes = classes[indices]
    else:
        final_boxes = np.array([])
        final_scores = np.array([])
        final_classes = np.array([])

    # Annotate the image
    image_with_boxes = image_np.copy()
    if final_boxes.size > 0:
        for box in final_boxes:
            x_center, y_center, width, height = box
            x1 = int((x_center - width / 2) * image_width)
            y1 = int((y_center - height / 2) * image_height)
            x2 = int((x_center + width / 2) * image_width)
            y2 = int((y_center + height / 2) * image_height)
            cv2.rectangle(image_with_boxes, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Save annotated image
    cv2.imwrite(image_save_path, cv2.cvtColor(image_with_boxes, cv2.COLOR_RGB2BGR))

    # Save annotations in YOLO format
    with open(annotation_save_path, 'w') as f:
        for box, cls in zip(final_boxes, final_classes):
            x_center, y_center, width, height = box
            line = f"{cls} {x_center} {y_center} {width} {height}\n"
            f.write(line)

    return final_boxes, image_with_boxes
```

```python
import numpy as np

def bounding_box_list(filename):
    # Initialize an empty list to hold the bounding boxes
    bounding_boxes = []

    # Read the data from the file
    with open(filename, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) == 5:
                # If class ID is present
                class_id = int(parts[0])
                coords = list(map(float, parts[1:]))
            elif len(parts) == 4:
                # If class ID is missing, assume 0
                class_id = 0
                coords = list(map(float, parts))
            else:
                continue  # Skip invalid lines

            bounding_boxes.append({
                'class_id': class_id,
                'x_center': coords[0],
                'y_center': coords[1],
                'width': coords[2],
                'height': coords[3]
            })
10    return bounding_boxes
```

```python
def yolo_to_corners(box):
    x_center = box['x_center']
    y_center = box['y_center']
    width = box['width']
    height = box['height']
    x_min = x_center - width / 2
    x_max = x_center + width / 2
    y_min = y_center - height / 2
    y_max = y_center + height / 2
    return x_min, y_min, x_max, y_max
```

```python
ef corners_to_yolo(x_min, y_min, x_max, y_max):
    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2
    width = x_max - x_min
    height = y_max - y_min
    return x_center, y_center, width, height
```

```python
def calculate_iou(box1, box2):
    x1_min, y1_min, x1_max, y1_max = yolo_to_corners(box1)
    x2_min, y2_min, x2_max, y2_max = yolo_to_corners(box2)

    # Intersection coordinates
    inter_x_min = max(x1_min, x2_min)
    inter_y_min = max(y1_min, y2_min)
    inter_x_max = min(x1_max, x2_max)
    inter_y_max = min(y1_max, y2_max)

    # Compute intersection area
    inter_width = max(0, inter_x_max - inter_x_min)
    inter_height = max(0, inter_y_max - inter_y_min)
    inter_area = inter_width * inter_height

    # Compute areas of the bounding boxes
    area1 = (x1_max - x1_min) * (y1_max - y1_min)
    area2 = (x2_max - x2_min) * (y2_max - y2_min)

    # Compute union area
    union_area = area1 + area2 - inter_area

    # Compute IoU
    iou = inter_area / union_area if union_area != 0 else 0
    return iou
```

```python
def merge_boxes(box1, box2):
    x1_min, y1_min, x1_max, y1_max = yolo_to_corners(box1)
    x2_min, y2_min, x2_max, y2_max = yolo_to_corners(box2)

    # Coordinates of the merged box
    merged_x_min = min(x1_min, x2_min)
    merged_y_min = min(y1_min, y2_min)
    merged_x_max = max(x1_max, x2_max)
    merged_y_max = max(y1_max, y2_max)

    # Convert back to YOLO format
    x_center, y_center, width, height = corners_to_yolo(
        merged_x_min, merged_y_min, merged_x_max, merged_y_max)

    return {
        'class_id': box1['class_id'],  # Assuming same class
        'x_center': x_center,
        'y_center': y_center,
        'width': width,
        'height': height
    }
```

```python
def merge_overlapping_boxes(boxes, iou_threshold=0.45):
    merged = True
    while merged:
        merged = False
        num_boxes = len(boxes)
        for i in range(num_boxes):
            box1 = boxes[i]
            for j in range(i + 1, num_boxes):
                box2 = boxes[j]
                iou = calculate_iou(box1, box2)
                if iou > iou_threshold:
                    # Merge the two boxes
                    new_box = merge_boxes(box1, box2)
                    # Remove the original boxes
                    boxes.pop(j)
                    boxes.pop(i)
                    # Add the new merged box
                    boxes.append(new_box)
                    merged = True
                    break  # Need to restart since list has changed
            if merged:
                break
    return boxes
```

```python
from PIL import Image

def convert_png_to_jpg(input_file: str, output_file: str) -> None:
    """
    Convert a PNG image to JPG format.

    Args:
        input_file (str): Path to the input PNG file
        output_file (str): Path where the output JPG file should be saved

    Raises:
        FileNotFoundError: If the input file doesn't exist
        OSError: If there's an error processing the image
    """
    try:
        # Open the PNG image
        with Image.open(input_file) as img:
            # Convert to RGB mode if necessary (in case of RGBA PNG)
            if img.mode in ('RGBA', 'LA'):
                background = Image.new('RGB', img.size, (255, 255, 255))
                background.paste(img, mask=img.split()[-1])
                img = background

            # Save as JPG
            img.save(output_file, 'JPEG', quality=95)

    except FileNotFoundError:
        raise FileNotFoundError(f"Input file not found: {input_file}")
    except Exception as e:
        raise OSError(f"Error converting image: {str(e)}")
```
